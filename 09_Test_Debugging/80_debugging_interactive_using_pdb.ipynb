`breakpoint()` 를 호출하면 거기서 대화형 pdb 디버거가 실행된다.

`locals` 내장 함수를 호출하면 모든 지역 변수 목록을 볼 수 있고

모듈을 임포트하거나 새로운 객체를 만들거나 help 내장 함수를 실행하거나, 프로그램의 일부를 변경할 수 도 있다.

다음 세 가지 명령은 실행 중인 프로그램을 관찰 할 때 아주 유용하다

- where: 현재 실행 중인 프고르맹의 호출 스택을 변경, 이 명령을 사용하면 실행 중인 프로그램의 현재 위치를 알 수 있고 
         breakpoint 트리거가 어떻게 발동 됬는지 알 수 있다.

- up: 실행 호출 스택에서 현재 관찰중인 함수를 호출한 쪽(위)으로 호출 스택 영역을 한 단계 이동해서 해당 함수의 지역 변수를 관찰 할 수 있게 해준다.
      up 대신 u 만 써도 된다

- down: 실행 호출 스택에서 한 수준 아래로 호출 스택 영역을 이동한다. Down대신 d만 써도 된다.

현재 상태를 본후에는 다음 다섯가지 디버거 명령을 사용해 프로그램 실행을 다양한 방식으로 제어할 수 있다.

- step: 프로그램을 다음 줄까지 실행한 다음 제어를 디버거로 돌려서 디버거 프롬프트를 표시한다. <p></p>
        소스 코드 다음줄에 함수를 호출하는 부분이 있다면 **해당 함수의 첫 줄에서 디버거로 제어가 돌아온다.**

- next: 프로그램을 다음 줄까지 실행한 다음 제어를 다음 제어를 디버거로 돌려서 디버거 프롬프트를 표시한다. <p></p>
        소스 코드의 다음 줄에 함수를 호출하는 바분이 있다면 **해당 함수에서 반환된 다음에 제어가 디버거로 돌아온다.**

- return: 현재 함수에서 반환될 때까지 프로그램을 실행한 후 제어가 디버거로 돌아온다.

- continue: 다음 중단점에 도달할 때까지 프로그램을 계속실행한다.

- quit: 디버거에서 나가면서 프로그램도 중단 시킨다. 문제의 원인을 찾았거나, ... 

디버거를 시작하는 방법으로 사후 디버깅 (post-mortem debugging)이 있다.

프로그램에 문제가 생겨 중단된 뒤에 프로그램을 디버깅 할 수 있다.

breakpoint 함수 호출을 어디에 추가해아할지 잘 모르는 경우 도움이 된다

`python3 -m pdf -c continue postmortem_breakpoint.py`
